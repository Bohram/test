Задание 2. Работа с выбросами

В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
# Рассмотрим выбросы по величине rectal_temperature

q1 = df_horse['rectal_temperature'].quantile(0.25)
q3 = df_horse['rectal_temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
print(f'{lower_bound} нижняя граница')
print(f'{upper_bound} верхняя граница')

df_rectal = df_horse[df_horse['rectal_temperature']>upper_bound]
df_rectal.head()
```

```python
# Рассмотрим выбросы по величине pulse

q1 = df_horse['pulse'].quantile(0.25)
q3 = df_horse['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
print(f'{lower_bound} нижняя граница')
print(f'{upper_bound} верхняя граница')

df_pulse = df_horse[df_horse['pulse']>upper_bound]
df_pulse.head()
```

Краткий вывод:
rectal_temperature - не сильно отличаются от среднего, чтобы вызывать подозрение.

pulse - Данные значения пульса выбиваются из общей картины, среднем пульс должен быть 20-25, при этом третий квартиль у нас 88, Пульс 150-184 в 2-3 раза больше. Нужно обратиться к специалисту для интерпритации, т.к. либо опечатка, либо показания пулься были получены при пограничном состоянии здоровья лошади (повышенная температура может свидетельствовать о заболевании), либо был испорчен прибор.

Задание 3. Работа с пропусками

Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
# Процентное отображение пропусков
for col in df_horse.columns:
    pct_missing = df_horse[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
#Заполняем пропуски медианным значением в зависимости от возраста и наличия операций
df['rectal_temperature'].fillna(df.groupby(['Age', 'surgery'])['rectal_temperature'].transform('median'), inplace=True)
df['temperature_of_extremities'].fillna(df.groupby(['Age', 'surgery'])['temperature_of_extremities'].transform('median'), inplace=True)
df['total_protein'].fillna(df.groupby(['Age', 'surgery'])['total_protein'].transform('median'), inplace=True)
df['pulse'].fillna(df.groupby(['Age', 'surgery'])['pulse'].transform('median'), inplace=True)
df_2 = df[['surgery', 'Age', 'Hospital_Number', 'rectal_temperature', 'pulse', 'temperature_of_extremities', 'total_protein', 'outcome']]
df_2.dropna(inplace=True)
df_2
```

```python
# Проверим на результат
for col in df_2.columns:
    pct_missing = df_2[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```
