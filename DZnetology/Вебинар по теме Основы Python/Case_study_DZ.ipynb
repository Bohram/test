Вопрос №1
Как критики относятся к спортивным играм?

H0 - Критики относятся к спортивным играм так же как и к другим играм, будет почти совпадать со средним.

H1 - Критики по спортивным играм, относятся кретичнее, а значит оценки будут ниже среднего.

Уровень значимости - 0.05

```python
a = 0.05
```

```python
df.describe() #Интересно посмотреть что в общем происходит
```

Critic_Score - Узнали среднее значение по всем играм = 7,2, а так же другие статистические показатели.

```python
df_sport = df[df.Genre == 'Sports'].Critic_Score  # Посмотрим среднию оценку по сортивным играм
df_sport.mean()
```

```python
mean_general = df.Critic_Score.mean()    # Зафиксируем среднию оценку по всем имграм
mean_general
```

```python
stats, p = st.ttest_1samp(df_sport, mean_general)
print('Statistics=%.3f, p=%.3f' % (stats, p))

if (stats < 0) & (p / 2 < a):
    print('Отвергаем H0, критики более кретичны к спортивным играм')
else:
    print('Не отвергаем Н0, критики относятся к спортивным играм так же как и к другим играм')
```

2) Критикам нравятся больше игры на PC или на PS4?

H0 - Критикам больше нравиться игры на PC

H1 - Критикам меньше нравиться игры на PС

Уровень значимости - 0.05

```python
df_PC = df[df.Platform == 'PC'].Critic_Score.dropna()
df_PS4 = df[df.Platform == 'PS4'].Critic_Score.dropna()
print("Средняя оценка игра на PC:",df_PC.mean(),"Средняя оценка игра на PS4:",df_PS4.mean())
```

```python
PC = df.loc[df['Platform'] == 'PC']
PS4 = df.loc[df['Platform'] == 'PS4']
```

```python
stats, p = st.ttest_ind(PC['Critic_Score'], PS4['Critic_Score'], equal_var=False)
print('Statistics=%.3f, p=%.3f' % (stats, p))

if (stats < 0) & (p / 2 < a):
    print('Отвергаем H0, Критикам меньше нравиться игры на PС')
else:
    print('Не отвергаем H0, Критикам больше нравиться игры на PC')
```

H0 - Критики относятся к стрелялкам так же как и к другим играм, будет почти совпадать со средним.

H1 - Критикам больше нравяться стрелялки.

```python
Shooter = df.loc[df['Genre'] == 'Shooter']
Strategy = df.loc[df['Genre'] == 'Strategy']
```

```python
stat, p = st.ttest_ind(Shooter['Critic_Score'], Strategy['Critic_Score'], equal_var=False)
print('Statistics=%.3f, p=%.3f' % (stat, p))

if (p < a):
    print('Отвергаем H0, критикам больше нравяться стрелялки.')
else:
    print('Не отвергаем H0, критики относятся к стрелялкам так же как и к другим играм,')
```
